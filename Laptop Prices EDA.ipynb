```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Price_euros', hue='Company', bins=10, kde=True, multiple="stack")

# Add labels
plt.xlabel('Price (Euros)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Encode categorical variables to numeric for regplot
data['Product'] = data['Product'].astype('category').cat.codes
data['CPU_model'] = data['CPU_model'].astype('category').cat.codes

# Create the regression plot
plt.figure(figsize=(10, 10))
sns.regplot(y='Product', x='Price_euros', data=data)
sns.regplot(y='CPU_model', x='Price_euros', data=data, color='orange')

# Adding labels and title
plt.xlabel('Product and CPU Model')
plt.ylabel('Price in Euros')
plt.xticks(ticks=data['Product'].unique(), labels=data['Product'].unique(), rotation=45)
plt.grid(True)
plt.show()
```
